# Analyse des erreurs de géoréférencement — multi-outages

Charge les LAS `rmse_*.las` (champ `e3d`) produits par `rmse_streaming.py`  
pour **plusieurs outages simultanément**, avec bornes et exclusions indépendantes.

**Dépendances** : `laspy`, `numpy`, `pandas`, `matplotlib`

```python
from pathlib import Path
import numpy as np
import pandas as pd
import laspy
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
plt.rcParams.update({
    'figure.dpi': 130,
    'font.size': 10,
    'axes.grid': True,
    'grid.alpha': 0.35,
})
```

## ① Configuration — un bloc par outage

Ajoute/supprime des entrées dans `OUTAGES`.  
Chaque outage est un `dict` avec :
- `id` : label affiché
- `output_dir` : dossier racine contenant les sous-dossiers par méthode
- `methods` : liste de noms de méthodes (ou `None` pour détecter auto)
- `t_min` / `t_max` : bornes du plot en GPS time (SOW)
- `exclusion_zones` : liste de tuples `(t_a, t_b)` à griser
- `imu` : label de l'IMU utilisé (affiché dans les titres)

```python
BASE = Path("/media/b085164/LaCie/2026spring_RD/CALIB_26_02_25/georef_errors")
BASE_EL = Path("/media/b085164/Elements/CALIB_26_02_25")

OUTAGES = [
    {
        "id":    "Outage 1",
        "imu":   "APX",
        "output_dir": BASE / "outage_1" / "APX",
        "methods": ["OUTAGE", "F2B", "COMBINED"],
        "t_min": 305200.0,
        "t_max": 305620.0,
        "exclusion_zones": [],   # à remplir si besoin
    },
    {
        "id":    "Outage 2",
        "imu":   "APX",
        "output_dir": BASE / "outage_2" / "APX",
        "methods": ["OUTAGE", "F2B", "COMBINED"],
        "t_min": 305645.0,
        "t_max": 306120.0,
        "exclusion_zones": [],
    },
    {
        "id":    "Outage 3",
        "imu":   "APX",
        "output_dir": BASE / "outage_3" / "APX",
        "methods": ["OUTAGE", "F2B", "COMBINED"],
        "t_min": 306290.0,
        "t_max": 306645.0,
        "exclusion_zones": [(306405.0, 306560.0)],
    },
]

# ── Paramètres globaux ────────────────────────────────────────────────────────
THRESHOLD_LINES = [0.10, 0.50, 1.00]   # traits horizontaux (m)
SMOOTH_WINDOW   = 50                    # lissage (pts), 1 = désactivé

METHOD_COLORS = {
    "outage_traj": "#e74c3c",   # rouge  — INS-only
    "F2B":         "#f39c12",   # orange
    "S2S":         "#3498db",   # bleu
    "COMBINED":    "#27ae60",   # vert
}
_fallback_colors = plt.rcParams['axes.prop_cycle'].by_key()['color']

def method_color(name, idx=0):
    return METHOD_COLORS.get(name, _fallback_colors[idx % len(_fallback_colors)])
```

## ② Fonctions utilitaires

```python
def load_las_e3d(las_path: Path):
    """Retourne (t, e3d) triés depuis un LAS rmse_streaming."""
    with laspy.open(las_path) as f:
        pts = f.read()
    t   = np.asarray(pts.gps_time, dtype=np.float64)
    e3d = np.asarray(pts['e3d'],   dtype=np.float32)
    order = np.argsort(t)
    return t[order], e3d[order]


def discover_and_load(output_dir: Path, methods=None):
    """
    Charge tous les LAS rmse_*.las dans output_dir/<method>/.
    Retourne :
      data        : {method: (t, e3d)}
      file_index  : {method: [Path, ...]}
    """
    if methods is None:
        methods = sorted(d.name for d in output_dir.iterdir() if d.is_dir())

    data, file_index = {}, {}
    for m in methods:
        mdir = output_dir / m
        if not mdir.exists():
            print(f"  ⚠ {m}: dossier introuvable ({mdir})")
            continue
        files = sorted(mdir.glob("rmse_*.las"))
        if not files:
            print(f"  ⚠ {m}: aucun rmse_*.las")
            continue
        file_index[m] = files
        ts, es = [], []
        for f in files:
            try:
                t, e = load_las_e3d(f)
                ts.append(t); es.append(e)
            except Exception as ex:
                print(f"  ✗ {f.name}: {ex}")
        if ts:
            t_cat = np.concatenate(ts)
            e_cat = np.concatenate(es)
            order = np.argsort(t_cat)
            data[m] = (t_cat[order], e_cat[order])
    return data, file_index


def compute_metrics(t, e, t_min=None, t_max=None, excl_zones=None):
    mask = np.ones(len(t), dtype=bool)
    if t_min is not None: mask &= (t >= t_min)
    if t_max is not None: mask &= (t <= t_max)
    for za, zb in (excl_zones or []):
        mask &= ~((t >= za) & (t <= zb))
    e = e[mask]
    nan = {k: np.nan for k in ['N','RMSE','mean','Q50','Q75','Q90','Q95','Q99',
                                'max','%>0.10m','%>0.50m','%>1.00m']}
    if len(e) == 0: return nan
    return {
        'N':        len(e),
        'RMSE':     float(np.sqrt(np.mean(e**2))),
        'mean':     float(np.mean(e)),
        'Q50':      float(np.percentile(e, 50)),
        'Q75':      float(np.percentile(e, 75)),
        'Q90':      float(np.percentile(e, 90)),
        'Q95':      float(np.percentile(e, 95)),
        'Q99':      float(np.percentile(e, 99)),
        'max':      float(np.max(e)),
        '%>0.10m':  float(np.mean(e > 0.10) * 100),
        '%>0.50m':  float(np.mean(e > 0.50) * 100),
        '%>1.00m':  float(np.mean(e > 1.00) * 100),
    }


def smooth(arr, w):
    if w <= 1 or len(arr) < w: return arr
    return np.convolve(arr, np.ones(w)/w, mode='same')
```

## ③ Chargement de tous les outages

```python
# Résultat : liste de dicts enrichis avec data + file_index
outages = []
for cfg in OUTAGES:
    print(f"\n{'─'*60}")
    print(f"  {cfg['id']} — {cfg['imu']}  [{cfg['t_min']:.0f} – {cfg['t_max']:.0f}]")
    d, fi = discover_and_load(cfg['output_dir'], cfg.get('methods'))
    for m, (t, e) in d.items():
        print(f"  ✓ {m:20s}  {len(t):>10,} pts  t=[{t.min():.1f}, {t.max():.1f}]")
    outages.append({**cfg, 'data': d, 'file_index': fi})
```

## ④ Tableau de métriques — tous outages

Index : `(outage_id, imu, method)` — colonnes principales : RMSE, Q50, Q95, Q99, max, %>0.50m, %>1.00m.

```python
rows = []
for o in outages:
    for method, (t, e) in o['data'].items():
        m = compute_metrics(t, e, o['t_min'], o['t_max'], o['exclusion_zones'])
        rows.append({'outage': o['id'], 'imu': o['imu'], 'method': method, **m})

df_all = pd.DataFrame(rows).set_index(['outage', 'imu', 'method'])

fmt = {
    'N':        '{:,.0f}',
    'RMSE':     '{:.4f} m',
    'mean':     '{:.4f} m',
    'Q50':      '{:.4f} m',
    'Q75':      '{:.4f} m',
    'Q90':      '{:.4f} m',
    'Q95':      '{:.4f} m',
    'Q99':      '{:.4f} m',
    'max':      '{:.4f} m',
    '%>0.10m':  '{:.1f} %',
    '%>0.50m':  '{:.1f} %',
    '%>1.00m':  '{:.1f} %',
}

display(
    df_all.style
    .format(fmt, na_rep='—')
    .highlight_min(subset=['RMSE','Q95','Q99'], color='#d4edda', axis=0)
    .highlight_max(subset=['RMSE','Q95','Q99'], color='#f8d7da', axis=0)
    .set_caption("Métriques d'erreur 3D — tous outages (exclusions appliquées)")
    .set_table_styles([{'selector':'th','props':'background:#f0f0f0;font-weight:bold'}])
)
```

## ⑤ Plot erreur 3D vs temps GPS — un panneau par outage

Chaque panneau affiche toutes les méthodes de cet outage sur le même axe.

```python
n_out = len(outages)
fig, axes = plt.subplots(n_out, 1, figsize=(15, 4.5 * n_out), squeeze=False)

for row_i, o in enumerate(outages):
    ax = axes[row_i, 0]
    t_min, t_max = o['t_min'], o['t_max']
    excl = o['exclusion_zones']

    for m_i, (method, (t, e)) in enumerate(o['data'].items()):
        color = method_color(method, m_i)

        mask = np.ones(len(t), dtype=bool)
        if t_min is not None: mask &= (t >= t_min)
        if t_max is not None: mask &= (t <= t_max)
        t_p, e_p = t[mask], e[mask]

        # points bruts (très transparents)
        ax.scatter(t_p, e_p, s=0.8, alpha=0.15, color=color, linewidths=0)

        # courbe lissée
        e_sm = smooth(e_p, SMOOTH_WINDOW)
        mets = compute_metrics(t, e, t_min, t_max, excl)
        label = f"{method}  RMSE={mets['RMSE']:.3f} m  Q95={mets['Q95']:.3f} m"
        ax.plot(t_p, e_sm, color=color, lw=1.8, alpha=0.9, label=label)

        # RMSE et Q95 en tirets
        ax.axhline(mets['RMSE'], color=color, lw=0.9, ls='--', alpha=0.7)
        ax.axhline(mets['Q95'],  color=color, lw=0.9, ls=':',  alpha=0.7)

    # seuils horizontaux
    for thr in THRESHOLD_LINES:
        ax.axhline(thr, color='gray', lw=0.6, alpha=0.4)
        ax.text(t_min or ax.get_xlim()[0], thr + 0.01,
                f'{thr:.2f} m', fontsize=7, color='gray', va='bottom')

    # zones exclues
    for za, zb in excl:
        ax.axvspan(za, zb, color='gray', alpha=0.13, zorder=0)
        ax.text((za+zb)/2, ax.get_ylim()[1]*0.92 if ax.get_ylim()[1]>0 else 1,
                'exclu', ha='center', fontsize=7, color='gray')

    # bornes outage
    ax.axvline(t_min, color='black', lw=1.0, ls='--', alpha=0.55, label='début outage')
    ax.axvline(t_max, color='black', lw=1.0, ls='-.', alpha=0.55, label='fin outage')

    dur = t_max - t_min
    ax.set_title(f"{o['id']} — {o['imu']}  |  [{t_min:.0f} – {t_max:.0f}]  ({dur:.0f} s)",
                 fontsize=11, fontweight='bold')
    ax.set_ylabel('Erreur 3D (m)', fontsize=9)
    ax.legend(loc='upper left', fontsize=8, framealpha=0.88, markerscale=5)
    ax.yaxis.set_minor_locator(mticker.AutoMinorLocator())

axes[-1, 0].set_xlabel('Temps GPS — secondes of week (s)', fontsize=10)
fig.suptitle('Erreur de géoréférencement 3D vs. ODyN référence', fontsize=13, y=1.005)
fig.tight_layout()
plt.show()
```

## ⑥ Plot grille : outage × méthode (un sous-plot par combinaison)

Vue détaillée pour comparer chaque méthode indépendamment à travers les outages.

```python
# Toutes les méthodes présentes dans au moins un outage
all_methods = []
for o in outages:
    for m in o['data']:
        if m not in all_methods:
            all_methods.append(m)

n_rows = len(outages)
n_cols = len(all_methods)
fig, axes = plt.subplots(n_rows, n_cols,
                         figsize=(5 * n_cols, 3.8 * n_rows),
                         sharex='row', sharey='row',
                         squeeze=False)

for ri, o in enumerate(outages):
    t_min, t_max = o['t_min'], o['t_max']
    excl = o['exclusion_zones']

    for ci, method in enumerate(all_methods):
        ax = axes[ri, ci]
        color = method_color(method, ci)

        if method not in o['data']:
            ax.text(0.5, 0.5, 'n/a', ha='center', va='center',
                    transform=ax.transAxes, color='gray', fontsize=14)
            ax.set_visible(True)
            if ri == 0: ax.set_title(method, fontsize=10, fontweight='bold', color='gray')
            continue

        t, e = o['data'][method]
        mask = np.ones(len(t), dtype=bool)
        if t_min is not None: mask &= (t >= t_min)
        if t_max is not None: mask &= (t <= t_max)
        t_p, e_p = t[mask], e[mask]

        ax.scatter(t_p, e_p, s=0.6, alpha=0.12, color=color, linewidths=0)
        e_sm = smooth(e_p, SMOOTH_WINDOW)
        ax.plot(t_p, e_sm, color=color, lw=1.5)

        mets = compute_metrics(t, e, t_min, t_max, excl)
        ax.axhline(mets['RMSE'], color=color, lw=1.0, ls='--', alpha=0.8,
                   label=f"RMSE={mets['RMSE']:.3f} m")
        ax.axhline(mets['Q95'],  color=color, lw=1.0, ls=':',  alpha=0.8,
                   label=f"Q95={mets['Q95']:.3f} m")

        for thr in THRESHOLD_LINES:
            ax.axhline(thr, color='gray', lw=0.5, alpha=0.35)

        for za, zb in excl:
            ax.axvspan(za, zb, color='gray', alpha=0.12, zorder=0)

        ax.legend(fontsize=7, framealpha=0.85, loc='upper left')

        if ri == 0:
            ax.set_title(method, fontsize=10, fontweight='bold', color=color)
        if ci == 0:
            ax.set_ylabel(f"{o['id']}\n{o['imu']}\nerreur 3D (m)", fontsize=8)
        if ri == n_rows - 1:
            ax.set_xlabel('GPS time (SOW, s)', fontsize=8)

fig.suptitle('Vue grille : outage × méthode', fontsize=12, y=1.01)
fig.tight_layout()
plt.show()
```

## ⑦ Histogrammes cumulés (CDF) — comparaison inter-outage par méthode

```python
# Un subplot par méthode, CDF de chaque outage en surimpression
fig, axes = plt.subplots(1, len(all_methods),
                         figsize=(5 * len(all_methods), 4),
                         sharey=True, squeeze=False)

outage_linestyles = ['-', '--', '-.', ':']

for ci, method in enumerate(all_methods):
    ax = axes[0, ci]
    color = method_color(method, ci)

    for oi, o in enumerate(outages):
        if method not in o['data']:
            continue
        t, e = o['data'][method]
        mets = compute_metrics(t, e, o['t_min'], o['t_max'], o['exclusion_zones'])
        # Filtre pour la CDF
        mask = np.ones(len(t), dtype=bool)
        if o['t_min']: mask &= (t >= o['t_min'])
        if o['t_max']: mask &= (t <= o['t_max'])
        for za, zb in o['exclusion_zones']:
            mask &= ~((t >= za) & (t <= zb))
        e_f = np.sort(e[mask])
        cdf = np.arange(1, len(e_f)+1) / len(e_f)

        ls = outage_linestyles[oi % len(outage_linestyles)]
        label = f"{o['id']} ({o['imu']})  RMSE={mets['RMSE']:.3f} m"
        ax.plot(e_f, cdf, color=color, ls=ls, lw=1.6, label=label)

    for thr in THRESHOLD_LINES:
        ax.axvline(thr, color='gray', lw=0.7, ls='--', alpha=0.5)
        ax.text(thr+0.005, 0.02, f'{thr:.2f} m', fontsize=7, color='gray', rotation=90, va='bottom')

    ax.set_xlim(left=0)
    ax.set_ylim(0, 1)
    ax.set_xlabel('Erreur 3D (m)', fontsize=9)
    ax.set_title(method, fontsize=10, fontweight='bold', color=color)
    ax.legend(fontsize=7.5, framealpha=0.9, loc='lower right')

axes[0, 0].set_ylabel('CDF', fontsize=9)
fig.suptitle('CDF de l\'erreur 3D par méthode — comparaison inter-outages', fontsize=11)
fig.tight_layout()
plt.show()
```

## ⑨ Stats par scan individuel — par outage

```python
scan_rows = []
for o in outages:
    for method, files in o['file_index'].items():
        for las_file in sorted(files):
            try:
                t, e = load_las_e3d(las_file)
            except Exception as ex:
                print(f"✗ {las_file.name}: {ex}"); continue
            try:
                scan_id = int(las_file.stem.split('scan')[-1])
            except ValueError:
                scan_id = -1
            m = compute_metrics(t, e, o['t_min'], o['t_max'], o['exclusion_zones'])
            m.update({'outage': o['id'], 'imu': o['imu'],
                      'method': method, 'scan_id': scan_id,
                      't_start': float(t.min()) if len(t) else np.nan,
                      't_end':   float(t.max()) if len(t) else np.nan})
            scan_rows.append(m)

df_scans = (pd.DataFrame(scan_rows)
            .set_index(['outage','imu','method','scan_id'])
            .sort_index())

cols_show = ['N','t_start','t_end','RMSE','Q50','Q95','Q99','max','%>0.50m','%>1.00m']
fmt_scan  = {c: ('{:,.0f}' if c=='N' else
                 '{:.1f}'  if c in ('t_start','t_end') else
                 '{:.1f} %' if c.startswith('%') else '{:.4f}')
             for c in cols_show}

display(
    df_scans[cols_show].style
    .format(fmt_scan, na_rep='—')
    .highlight_max(subset=['RMSE','Q95'], color='#f8d7da')
    .highlight_min(subset=['RMSE','Q95'], color='#d4edda')
    .set_caption('Stats par scan individuel — tous outages')
)
```
